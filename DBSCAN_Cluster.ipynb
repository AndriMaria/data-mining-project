```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt


df = pd.read_csv('S006.csv')


scaler = StandardScaler()
df1 = df[['label', 'back_x', 'back_y', 'back_z', 'thigh_x', 'thigh_y', 'thigh_z']]  # Select relevant features
df1_scaled = scaler.fit_transform(df1) 



sampling_frequency = 8 #η συχνοτητα δειγματοληψιας - μετρησεις ανα δευτερόλεπτο
time_interval = 2
#Διαδικασια αφαιρεσης γραμμων
rows_to_remove = range( 8, time_interval // sampling_frequency)
data_reduced = df1.drop(rows_to_remove).reset_index(drop=True)

dbscan_model = DBSCAN(eps=0.5, min_samples=5).fit(data_reduced)


df['cluster'] = dbscan_model.labels_


plt.figure(figsize=(10, 6))
plt.scatter(['label'],['back_x'], c=df['cluster'], cmap='dbscan_cluster', s=10)
plt.title('DBSCAN Clustering of Back_X vs Label')
plt.xlabel('label')
plt.ylabel('back_x')
plt.show()

```
