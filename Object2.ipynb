```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from seaborn import barplot

sns.set(color_codes=True)

Object2 = pd.read_csv('S008.csv')

print(Object2.columns)
print(Object2.info())
print(Object2.describe())
print(Object2.isnull().sum())
```

```python
Object2["label"].value_counts().plot(kind='bar',title='Συμμετέχων 2(S008)')
```

```python
Object2['back_x'].hist()
plt.show()
fig = plt.figure()
Object2['back_y'].hist()
plt.show()
fig1 = plt.figure()
Object2['back_z'].hist()
plt.show()
fig3 = plt.figure()
Object2['thigh_x'].hist()
plt.show()
fig4 = plt.figure()
Object2['thigh_y'].hist()
plt.show()
fig5 = plt.figure()
Object2['thigh_z'].hist()
plt.show()
fig6 = plt.figure()


ax = plt.axes(projection='3d')


z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)


ax.plot3D(Object2['back_x'], Object2['back_y'], Object2['back_z'])
ax.set_title('Back_8')
plt.show()




ax1 = plt.axes(projection='3d')


z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax1.plot3D(Object2['thigh_x'], Object2['thigh_y'],Object2['thigh_z'])
ax1.set_title('Thigh_8')
plt.show()
```

```python

df = pd.DataFrame(Object2)
value_counts = df['label'].value_counts()


plt.figure(figsize=(12, 6))  # Προσαρμογή μεγέθους γραφήματος
value_counts.plot(kind='bar')
plt.xlabel('Τιμές label')
plt.ylabel('Συχνότητα')
plt.title('Συμμετέχων 2(S009)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
