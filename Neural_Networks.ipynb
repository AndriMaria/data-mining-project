```python

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import timestamp
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#ПҶПҢПҒП„ПүПғО· ОҙОөОҙОҝОјОӯОҪПүОҪ
data8 = pd.read_csv('S023.csv')
#data8['timestamp'] = pd.to_datetime(data8['timestamp'])

#ОәОұОҪОҝОҪО№ОәОҝПҖОҝОҜО·ПғО· ОҙОөОҙОҝОјОӯОҪПүОҪ ОјОӯПғПү standardization
data8[['back_x', 'back_y', 'back_z', 'thigh_x', 'thigh_y', 'thigh_z']] = (
    data8[['back_x', 'back_y', 'back_z', 'thigh_x', 'thigh_y', 'thigh_z']]
    .apply(lambda x: (x - x.mean()) / x.std())
)

# ПҮПүПҒОҜО¶ОҝП…ОјОө П„Оұ ОҙОөОҙОҝОјОӯОҪОұ ПғП„Оұ ПҮОұПҒОұОәП„О·ПҒО№ПғП„О·ОәО¬ ОәОұО№ П„О№ПӮ ОөП„О№ОәОӯП„ОөПӮ
X = data8[['back_x', 'back_y', 'back_z', 'thigh_x', 'thigh_y', 'thigh_z']]
y = data8['label']

#ОіОҜОҪОөП„ОұО№ ОәПүОҙО№ОәОҝПҖОҝОҜО·ПғО· ПғП„О№ПӮ ОөП„О№ОәОӯП„ОөПӮ
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# ПҮПүПҒОҜПғОҝП…ОјОө ПғОө training set kai test set
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

# ПҶП„О№ОұПҮОҪОҝП…ОјОө ОөОҪОұ ПҖОҝО»П…ОөПҖОҜПҖОөОҙОҝ ОөПҖО№ПҖОөОҙОҝ ОјОө 4 dense ОөПҖОҜПҖОөОҙОұ
model = tf.keras.Sequential([#ОҙО№ОұОјПҢПҒПҶПүПғО· ОөПҖО№ПҖОӯОҙОҝП…
    tf.keras.layers.Input(shape=(X_train.shape[1],)), #ОөО№ПғОұОіПүОіО® ПҮОұПҒОұОәП„О·ПҒО№ПғП„О·ОәПҺОҪ
    tf.keras.layers.Dense(128, activation='relu'), #ПҖПҒПҺП„Оҝ ОөПҖОҜПҖОөОҙОҝ
    tf.keras.layers.Dense(64, activation='relu'),  #ОҙОөП…П„ОөПҒОҝ
    tf.keras.layers.Dense(32, activation='relu'),   #П„ПҒОҜП„Оҝ
    tf.keras.layers.Dense(len(label_encoder.classes_), activation='softmax') #ОөО№ПғО¬ОіОҝП…ОјОө П„О·ПӮ ОәПүОҙО№ОәОҝПҖОҝО№О·ОјОӯОҪОөПӮ ОөП„О№ОәОӯП„ОөПӮ/ОәО»О¬ПғОөО№ПӮ-П„ОӯП„ОұПҒП„Оҝ ОөПҖОҜПҖОөОҙОҝ
])

#ОәО¬ОҪОҝП…ОјОө complile П„Оҝ ОјОҝОҪП„ОӯО»Оҝ
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

#О•ОәПҖОұОҜОҙОөП…ПғО· ОјОҝОҪП„ОӯО»ОҝП… ОјОө 20 epochs ОәОұО№ validation split 20%
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2)


# ОөОјПҶО¬ОҪО№ПғО· ОұОәПҒОҜОІОөО№Оұ ОәОұО№ ОұПҖПҺО»ОөО№ОұПӮ ОҙОөО№ОіОјО¬П„ПүОҪ ПғП„Оҝ test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')

```
