```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


df = pd.read_csv('S006.csv')

plt.scatter(df.label,df['back_y'])
plt.xlabel('label')
plt.ylabel('back_y')
```

```python
km = KMeans(n_clusters=12)
y_predicted = km.fit_predict(df[['label','back_y']])


df['cluster']=y_predicted


df1 = df[df.cluster==1]
df2 = df[df.cluster==2]
plt.scatter(df1.label, df1['back_y'], color='green')
plt.scatter(df2.label, df2['back_y'], color='red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='blue',marker='*',label='centroid')
plt.xlabel('label')
plt.ylabel('back_y')
plt.legend()

```
